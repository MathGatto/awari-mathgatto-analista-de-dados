###**Exercício 1**

Usando o **Pandas** e a **base_rh.csv**, escreva um código que retorne o nome, o nível_de_carreira e a localidade de todos os profissionais que possuem nível de carreira = 'IC4' e que são da localidade de São Paulo.

```python
# Filtrando nível de carreira e localidade
filtro = (df["nivel_de_carreira"] == 'IC4') & (df['localidade'] == 'São Paulo')
df_exercicio1 = df[filtro]

# Definindo as colunas que serão mostradas
dados = ['nome', 'nivel_de_carreira', 'localidade']

# Mostrando o Resultado
df_exercicio1[dados].head()
```

###**Exercício 2**

Usando o **Pandas** e a **base_rh.csv**, escreva um código que retorne o nome dos funcionários em ordem crescente.

```python
# Ordenando por nome
df_exercicio2 = df.sort_values("nome")

# Mostrando o resultado
df_exercicio2.head(10)
```

###**Exercício 3**

Usando o **Pandas** e a **base_rh.csv**, escreva um código que retorne o nome e a data de nascimento dos profissionais que nasceram entre 1985 e 1990.

```python
# Filtrando por ano de nascimento
df_exercicio3 = df[(df['ano_nascimento'] >= 1985) & (df['ano_nascimento'] <= 1990)]

# Definindo as colunas que serão mostradas
dados = ['nome', 'nascimento', 'ano_nascimento']

# Mostrando o resultado
df_exercicio3[dados].head(10)
```

###**Exercício 4**

Usando o **Pandas** e a **base_rh.csv**, escreva um código que mostre quantos funcionários ativos existem por VPs.

```python
# Filtrando por status ativo
df_exercicio4 = df[(df['status'] == 'Ativo')]

# Agrupando por VP e fazendo um count para cada VP
resultado = df_exercicio4.groupby(['vp'])['nr'].count().reset_index(name='quantidade')

# Mostrando o resultado
resultado.head()
```

###**Exercício 5**

Usando o **Pandas** e a **base_rh.csv**, escreva um código que mostre o número de funcionários contratados por mês em 2022.

```python
# Realizar filtro do ano para 2022
df_exercicio5 = df[(df['ano_admissao'] == 2022)]

#Realizar agrupamento por mês e contar a quantidade
resultado = df_exercicio5.groupby(['mes_admissao'])['nr'].count().reset_index(name='quantidade')

#Exibir Resultado
resultado.head(12)
```

###**Exercício 6**

Usando o **Pandas** e a **base_rh.csv**, escreva um código que retorne o número de funcionários ativos na empresa.

```python
# Opção 1

# Contar os registros com status 'Ativo'
contagem_ativos = df['status'].value_counts()['Ativo']

# Exibir a contagem
print(contagem_ativos)
```

```python
# Opção 2

#Filtrar os registros ativos
df_exercicio6 = df[(df['status'] == 'Ativo')]

# Contar os registros já filtrados em ativos
resultado = len(df_exercicio6)

#Exibir Resultado
print('Número de funcionários ativos na empresa: ', resultado)
```
