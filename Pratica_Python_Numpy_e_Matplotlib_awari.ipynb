###**Exercício 3**
Criei um gráfico de linhas com eixo duplo, onde teremos duas informações

Linha 1: eixo x= nome, eixo y = crianças, cor = azul

Linha 2: eixo x= nome, eixo y = animais, cor = vermelho


```python
import matplotlib.pyplot as plt

# Dados
nome = ['Paulo','Daniel','Maria','Lucas','Tiago','Marina','Julia','Carla']
criancas = [2,0,0,3,2,1,4,0]
animais = [1,2,3,5,1,2,1,2]

# Criando um gráfico com título, rótulos e legenda
plt.plot(nome, criancas, label='Qtd de crianças', color='blue') #nome da legenda 1
plt.plot(nome, animais, label='Qtd de animais', color='red') #nome da legenda 2
plt.title('Gráfico de Linhas') #título
plt.xlabel('Eixo X') #eixo x
plt.ylabel('Eixo Y') #eixo y
plt.legend() #legenda
plt.show() #função para exibir o gráfico
```

###**Exercício 6**
Crie um gráfico de barras empilhadas agrupando VP e SEXO, contando o número de funcionários.

```python

df = pd.read_csv('/content/drive/MyDrive/MATHEUS GATTO - PESSOAL/Cursos/Awari/Awari - Formação Analista de Dados/base_rh.csv')

# Criar o gráfico de barras empilhadas
df.groupby(['vp', 'sexo']).size().unstack().plot(kind='bar', stacked=True)


# Configurar os rótulos dos eixos e o título do gráfico
plt.xlabel('VP')
plt.ylabel('Número de Funcionários')
plt.title('Número de Funcionários por VP e Sexo')


# Exibir o gráfico
plt.show()
```
